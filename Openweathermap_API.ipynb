```python
#using the weather API to find the weather data for a single city
#import the necessary libraries
import requests
import json
import pandas as pd
```

```python
#API key
API_key = '7b891adfc51b2719c62fc53cffdfe7fd'

#our city on interest
city_names = ['New York','Los Angeles','Chicago','Houston','Phoenix','Philadelphia','San Antonio','San Diego','Dallas','San Jose','Austin']

def API_request(city_names):

  name = []
  temperature = []
  max_temperature = []
  min_temperature = []
  humidity = []
  wind_speed = []
  description = []
  
  for city_name in city_names:

    #url or endpoint to which we are making our request, city_name and API key are the parameters to the API
    url = f'https://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_key}&units=imperial'

    #making a request to the API using the URL and the parameters
    r = requests.get(url)

    #checking if our request was successful
    if(r.status_code == 200):

      #by default, the api returns the data in json format. The .json() method returns the data
      data = r.json()
      
      #json.dumps() takes 3 parameters: the actualy json code, a boolean value which sorts the keys (sort_keys), and a value which specifies how much to indent
      formatted_json = json.dumps(data, sort_keys = True, indent = 4)
      #print(formatted_json)

      #extracting common weather information: temperature, humidity, and decription
      name.append(data['name'])
      temperature.append(data['main']['temp'])
      max_temperature.append(data['main']['temp_max'])
      min_temperature.append(data['main']['temp_min'])
      humidity.append(data['main']['humidity'])
      wind_speed.append(data['wind']['speed'])
      #weather is a list
      description.append(data['weather'][0]['description'])
    
    else:
      print('API request for',city_name,'was unsuccessful.','\n')

  #creating a dictionary and turning it into a dataframe
  weather_info = {'City Name' : name,
                  'Temperature' : temperature,
                  'Max Temperature' : max_temperature,
                  'Min Temperature' : min_temperature,
                  'Humidity' : humidity,
                  'Wind Speed' : wind_speed,
                  'Description' : description}
  weather = pd.DataFrame(weather_info)

  #creating a csv file
  weather.to_csv('topcities_weather.csv',index=False)
  return weather
```
