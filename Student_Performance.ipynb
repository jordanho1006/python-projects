```python
#changing the figure size
plt.rcParams['figure.figsize'] = [10 ,5]

#creating the list data for the bar graph
reading_scores = sum(student_performance['reading score'])/len(student_performance['reading score'])
writing_scores = sum(student_performance['writing score'])/len(student_performance['writing score'])
math_scores = sum(student_performance['math score'])/len(student_performance['math score'])
scores = [math_scores, reading_scores, writing_scores]
tests = ['Math', 'Reading', 'Writing']

#creating a list of colors to assign the bars to
colors = ['b','y','r']

#note the color parameter value
plt.bar(tests, scores, color = colors)
plt.title('Average Score for Each Test',fontsize=17)
plt.xlabel('Exam Type',fontsize=17)
plt.ylabel('Score',fontsize=17)
plt.show()
```

```python
#creating declarations for finding the average performance depending on parental level of education
highschool = highschool_counter = 0
some_college = some_college_counter = 0
associates_degree = associates_degree_counter = 0
some_highschool = some_highschool_counter = 0
bachelors_degree = bachelors_degree_counter = 0
masters_degree = masters_degree_counter = 0

#creating a loop to find which category each grade belongs in
for i in range(0,len(student_performance)):
  if student_performance['parental level of education'][i] == 'some college':
    some_college += student_performance['reading score'][i]
    some_college_counter += 1
  elif student_performance['parental level of education'][i] == "associate's degree":
    associates_degree += student_performance['reading score'][i]
    associates_degree_counter += 1
  elif student_performance['parental level of education'][i] == 'high school':
    highschool += student_performance['reading score'][i]
    highschool_counter += 1
  elif student_performance['parental level of education'][i] == 'some high school':
    some_highschool += student_performance['reading score'][i]
    some_highschool_counter += 1
  elif student_performance['parental level of education'][i] == "master's degree":
    masters_degree += student_performance['reading score'][i]
    masters_degree_counter += 1
  else:
    bachelors_degree += student_performance['reading score'][i]
    bachelors_degree_counter += 1

some_highschool = some_highschool/some_highschool_counter
highschool = highschool/highschool_counter
some_college = some_college/some_college_counter
associates_degree = associates_degree/associates_degree_counter
bachelors_degree = bachelors_degree/bachelors_degree_counter
masters_degree = masters_degree_counter

#creating the list for our bar graph
average_rgrades = [some_highschool, highschool, some_college, associates_degree, bachelors_degree, masters_degree]
parental_education = ['Some Highschool','Highschool','Some College',"Associate's Degree","Bachelor's Degree","Master's Degree"]

#creating the parameters for a horizontal bar graph
plt.barh( parental_education, average_rgrades, color = 'purple') 
plt.title('Parental Level of Education vs Reading Scores',fontsize=17)
plt.xlabel('Average Reading Test Scores',fontsize=15)
plt.ylabel('Parental Level of Education',fontsize=15)
plt.show()
```
